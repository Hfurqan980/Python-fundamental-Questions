### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here
# here we explain the code
### Chicken has 1 head and 2 legs
# dog has 1 head and 4 legs
# c + d = heads
# 2C + 4d = legs

# so we need to first find how many dogs we have and the remaining legs tells us the number of chicken
# c = heads - d
# expand the 2c + 4d = legs equation by putting the c = heads - d
# 2(heads - d) + 4d = legs -> 2heads - 2d + 4d = legs -> 2heads + 2d = legs -> 2d = (legs - 2heads)/2

heads = int(input("enter heads = "))
legs = int(input("enter legs = "))

dogs = (legs - 2*heads)//2
chicken = heads - dogs

print("Dogs ->",dogs,"chicken ->",chicken)
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Write your code here
n = int(input("Enter first natural number = "))

sum_of_square = 0
for i in range(1, n+1):
  sum_of_square += i*i

print ("The sum of square is = ",sum_of_square)

# One another way to find this
sum_of_squares = (n*(n+1)*(2*n+1))/6
print(sum_of_squares)
````
